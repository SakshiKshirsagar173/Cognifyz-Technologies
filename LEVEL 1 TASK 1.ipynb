TASK 1: DATA EXPLORATION AND PREPROCESSING

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Dataset Internship.csv")
```

```python
print("Dataset Shape: ", df.shape)
```

```python
df.info()
```

```python
missing_values=df.isnull().sum()
print("\n Missing values per columns\n",missing_values)
```

```python
df.fillna({'Cuisines':'Unknown'},inplace=True)
```

```python
plt.figure(figsize=(8,5))
```

```python
plt.title("Distribution of Aggregate Ratings")
plt.xlabel("Aggregate Rating")
plt.ylabel("Count")
```

```python
sns.histplot(df['Aggregate rating'], bins=10, kde=True,color='Green')
```

```python
ratings_counts=df['Aggregate rating'].value_counts().sort_index()
ratings_counts
```

```python
df.to_csv("Final_Data.csv", index=False)
print("Process complete")
```
